```python
from math import sin, cos, log, ceil
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

Setting-up the problem: constants and initial conditions

```python
# model parameters:
g = 9.81      # gravity in m s^{-2}
A = numpy.pi*(0.5**2)   # is the maximum cross sectional area of the rocket in m^2  
m_s = 50.  # weight of the rocket shell in kg
rho = 1.091    # is the average air density in kg / m^3
v_e = 325  # exhaust speed in m/s
C_D = 0.15  # drag coefficient

### set initial conditions ###
h0 = 0.0     # initial position of the rocket
v0 = 0.0 # initial velocity of the rocket
mp0 = 100.0  # initial weight of the rocket propellant
```

Time discretization

```python
T = 40.0                          # final time
dt = 0.1                           # time increment
N = int(T/dt) + 1                  # number of time-steps
t = numpy.linspace(0.0, T, N)      # time discretization
```

```python
h=numpy.zeros(N)
v=numpy.zeros(N)
mp=numpy.zeros(N)

h[0]=h0
v[0]=v0
mp[0]=mp0
```

```python
for i in range(N-1):
    h[i+1]=h[i]+dt*v[i]
    if mp[i]>0:
        v[i+1]=v[i]-dt*g+20.*dt*v_e/(m_s+mp[i])-0.5*dt*rho*abs(v[i])*v[i]*A*C_D/(m_s+mp[i])
        mp[i+1]=mp[i]-20.*dt
    if mp[i]==0:
        v[i+1]=v[i]-dt*g-0.5*dt*rho*abs(v[i])*v[i]*A*C_D/(m_s)
```

```python
# visualization of the attitude
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r't', fontsize=18)
plt.ylabel(r'h', fontsize=18)
plt.plot(t,h, 'r-', lw=2);
```

```python
# visualization of the velocity
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r't', fontsize=18)
plt.ylabel(r'v', fontsize=18)
plt.plot(t,v, 'g-', lw=2);
```
