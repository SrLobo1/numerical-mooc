# Sod's Shock Tube

```python
%matplotlib inline
import numpy 
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

### Initial conditions

```python
#Basic initial condition parameters
#defining grid size, time steps, etc...
nx = 81
nt = 51
dx = 20.0/(nx-1)
dt=0.0002
gamma=1.4

rho_l = 1.
rho_r = .125
rho=numpy.ones(nx)
rho[0:40]=rho_l
rho[40:]=rho_r

vel_l = 0.
vel_r = 0.
vel=numpy.ones(nx)
vel[0:40]=vel_l
vel[40:]=vel_r

p_l = 100000
p_r = 10000
p=numpy.ones(nx)
p[0:40]=p_l
p[40:]=p_r

x = numpy.linspace(0,20,nx)

u1=rho
u2=rho*vel
u3=p/(gamma-1.)+(rho*vel**2)/2
```

```python
plt.plot(x-10, rho, color='#003366', ls='-', lw=3)
plt.ylabel('rho (N/m^2)')
plt.xlabel('Distance (m)')
plt.ylim(-0.,1.1);
```

```python
plt.plot(x-10, vel, color='#009922', ls='-', lw=3)
plt.ylabel('vel (m/s)')
plt.xlabel('Distance (m)')
plt.ylim(-1,1);
```

```python
plt.plot(x-10, p, color='#833311', ls='-', lw=3)
plt.ylabel('p (N/m^2)')
plt.xlabel('Distance (m)')
plt.ylim(9000,110000);
```

### Iteration Model

```python
def computeF(u1,u2,u3,gamma):
   
    return u2,(u2**2)/u1+(gamma-1.)*(u3-0.5*(u2**2)/u1),(u3+(gamma-1.)*(u3-0.5*(u2**2)/u1))*u2/u1
```

```python
def richtmyer(u1,u2,u3, gamma, nt, dt, dx):
    #initialize our results array with dimensions nt by nx
    u1_n = numpy.zeros((nt,len(u1))) 
    u2_n = numpy.zeros((nt,len(u2))) 
    u3_n = numpy.zeros((nt,len(u3))) 
    
    u1_nm = numpy.zeros((nt,len(u1)-1)) 
    u2_nm = numpy.zeros((nt,len(u2)-1)) 
    u3_nm = numpy.zeros((nt,len(u3)-1)) 
    #copy the initial u array into each row of our new array
    u1_n[:,:] = u1.copy()
    u2_n[:,:] = u2.copy()
    u3_n[:,:] = u3.copy()
    
    for t in range(0,nt-1):
        [F1,F2,F3]=computeF(u1_n[t,:],u2_n[t,:],u3_n[t,:],gamma)
        u1_nm[t,0:]=0.5*(u1_n[t,1:]+u1_n[t,0:-1])-0.5*(dt/dx)*(F1[1:]-F1[0:-1])
        u2_nm[t,0:]=0.5*(u2_n[t,1:]+u2_n[t,0:-1])-0.5*(dt/dx)*(F2[1:]-F2[0:-1])
        u3_nm[t,0:]=0.5*(u3_n[t,1:]+u3_n[t,0:-1])-0.5*(dt/dx)*(F3[1:]-F3[0:-1])
        
        [F1m,F2m,F3m]=computeF(u1_nm[t,:],u2_nm[t,:],u3_nm[t,:],gamma)
        
        u1_n[t+1,1:-1]=u1_n[t,1:-1]-(dt/dx)*(F1m[1:]-F1m[0:-1])
        u2_n[t+1,1:-1]=u2_n[t,1:-1]-(dt/dx)*(F2m[1:]-F2m[0:-1])
        u3_n[t+1,1:-1]=u3_n[t,1:-1]-(dt/dx)*(F3m[1:]-F3m[0:-1])
    return u1_n,u2_n,u3_n   
```

```python
[u1_n,u2_n,u3_n]=richtmyer(u1,u2,u3, gamma,nt, dt, dx)
```

### Velocity at    t = 0.01 s

```python
velfin=u2_n[50,:]/u1_n[50,:]
plt.plot(x-10, velfin, color='#009922', ls='-', lw=3)
plt.ylabel('vel_final (m/s)')
plt.xlabel('Distance (m)')
plt.ylim(min(velfin),max(velfin));
```
